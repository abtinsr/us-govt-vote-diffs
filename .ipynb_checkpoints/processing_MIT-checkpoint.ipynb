```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
# https://dataverse.harvard.edu/dataverse/medsl_election_returns
pres = pd.read_csv('~/DN-Python/proj_US_election_results/data_mit/1976-2020-president.csv',
                   encoding='unicode_escape')
senate = pd.read_csv('~/DN-Python/proj_US_election_results/data_mit/1976-2020-senate.csv',
                     encoding='unicode_escape')
house = pd.read_csv('~/DN-Python/proj_US_election_results/data_mit/1976-2018-house3.csv',
                    encoding='unicode_escape')
```

```python
house.head(5)
```

```python
house = (
    house
    .rename(columns={'party': 'party_simplified'})
)
```

```python
house.year.unique()
```

```python
def house_votes_aggregated():

    df_parties = pd.DataFrame()

    years = house.year.unique()
    states = house.state.unique()

    for yr in years:
        for st in states:

            idx = np.where((house['state']==st) & (house['year']==yr))
            df = house.loc[idx]

            TOTAL_VOTES = df.groupby('state').aggregate({'candidatevotes':'sum'}).candidatevotes.values[0]

            df = pd.DataFrame(df.groupby(['state', 'party_simplified']).aggregate({'candidatevotes':'sum'}).reset_index())
            df['house_share_of_votes'] = round(df['candidatevotes']/TOTAL_VOTES,3)
            df['year'] = yr

            df_parties = df_parties.append(df)
    
    df_parties = df_parties.drop(['candidatevotes'], axis=1)
    
    return(df_parties)
```

```python
house_shares = house_votes_aggregated()
house_shares
```

#### Aggregate votes per state and party

```python
def votes_by_state_party_year(data, label):
    
    df = data
    
    df = (
        df.groupby(['year', 'state', 'party_simplified'])
        .aggregate({'candidatevotes':'sum', 'totalvotes':'mean'})
        .reset_index()
    )
    
    df[f'{label}_share_of_votes'] = round(df['candidatevotes']/df['totalvotes'],3)
    
    df = df[['year', 'state', 'party_simplified', f'{label}_share_of_votes']]
    
    return(df)
```

```python
# The function above does not account for the special election that occur in the Senate.
# It will aggregate votes that are not part of the same election.. 
# So, we need to create a special way of calculating Senate races.

def special_calc(data, label): 
    
    df = pd.DataFrame()
    
    for i in [True, False]:
        selected_races = data[data['special']==i]
        selected_races = votes_by_state_party_year(data=selected_races, label=label)
        df = df.append(selected_races)
    
    return(df)
```

```python
senate_shares = special_calc(data=senate, label='senate')
senate_shares.head(5)
```

```python
pres_shares = votes_by_state_party_year(data=pres, label='pres')
pres_shares.head(5)
```

#### Merge Presidential, Senate and House races

```python
def merge_races():
    
    vote_shares = pd.merge(pres_shares, senate_shares, how='inner', on=['year', 'state', 'party_simplified'])
    vote_shares = pd.merge(vote_shares, house_shares, how='inner', on=['year', 'state', 'party_simplified'])
    vote_shares['pres_to_senate_difference'] = vote_shares['pres_share_of_votes']-vote_shares['senate_share_of_votes']
    vote_shares['pres_to_house_difference'] = vote_shares['pres_share_of_votes']-vote_shares['house_share_of_votes']
    return(vote_shares)
```

```python
vote_shares = merge_races()
vote_shares.head(10)
```

```python
vote_shares.year.unique()
```

#### Visualise the difference in outcomes for certain states over time

```python
def visualiseFederalLevel(col, chamber):
    idx = np.where((vote_shares['party_simplified']=='REPUBLICAN') | (vote_shares['party_simplified']=='DEMOCRAT'))
    viz = vote_shares.loc[idx]
    viz = viz[['year', 'state', 'party_simplified', col]]

    plt.figure(figsize=(10,5))

    ax = sns.lineplot(data=viz, 
                      x='year',
                      y=col,
                      hue='party_simplified', 
                      palette=['Navy', 'Red'])

    plt.ylabel(f'Presidential race over/underperformance (versus {chamber})')
    plt.xlabel('Year')
    plt.title(f'How much better/worse do the major parties perform in Presidential than {chamber} races?')
    
    plt.axhline(y=0, color='black', linestyle='--', linewidth=1)

    vals = ax.get_yticks()
    ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
    ax.set_xticks(viz.year.unique()) # https://stackoverflow.com/questions/21910986/why-set-xticks-doesnt-set-the-labels-of-ticks/21911813#21911813
```

```python
visualiseFederalLevel(col='pres_to_senate_difference', chamber='Senate')
```

```python
def visualiseStateLevel(col, chamber):
    
    idx = np.where((vote_shares['party_simplified']=='REPUBLICAN') | (vote_shares['party_simplified']=='DEMOCRAT'))
    viz = vote_shares.loc[idx]
    viz = viz[['year', 'state', 'party_simplified', col]]

    plt.figure(figsize=(10,10))

    g = sns.relplot(data=viz, 
                    x='year',
                    y=col,
                    hue='party_simplified',
                    col='state',
                    kind='line',
                    col_wrap=2, 
                    palette=['Navy', 'Red'])

    (g.map(plt.axhline, y=0, color=".7", dashes=(2, 1), zorder=0)
      .set_axis_labels("Year", f"Presidential minus {chamber} votes (%)")
    )

    for ax in g.axes.flat: # Do this to add the axes to all graphs.
        ax.yaxis.set_major_formatter(PercentFormatter(1))
    
    ax.set_xticks(viz.year.unique())

    # https://stackoverflow.com/questions/54209895/seaborn-relplot-how-to-control-the-location-of-the-legend-and-add-title
    leg = g._legend
    leg.set_bbox_to_anchor([1, 1]);  # coordinates of lower left of bounding box
    #leg._loc = 2  # if required you can set the loc

    plt.show();
```

```python
visualiseStateLevel(col='pres_to_senate_difference', chamber='Senate');
```

```python
def visualiseVotes(party):
    idx = np.where((vote_shares['party_simplified']==party)) #| (vote_shares['party']=='democrat')
    viz = vote_shares.loc[idx]
    viz = viz[['year', 'state', 'party_simplified', 'pres_share_of_votes', 'senate_share_of_votes', 'house_share_of_votes']]

    viz.rename(columns={'pres_share_of_votes': 'presidential', 
                        'senate_share_of_votes': 'senate',
                        'house_share_of_votes': 'house'}, inplace=True)

    viz = viz.melt(id_vars=['year', 'state', 'party_simplified'], 
                  value_vars=['presidential', 'senate', 'house'],
                  var_name='race', value_name='share_of_votes')
    
    
    plt.figure(figsize=(10,5))

    ax = sns.lineplot(data=viz, 
                      x='year',
                      y='share_of_votes',
                      hue='race',
                      #style='party',
                      palette=['Purple', 'Orange', 'Green'])

    plt.ylabel(f'% Votes to {party} party')
    plt.xlabel('Year')
    plt.title(f'Votes to {party} party by political race and year')

    vals = ax.get_yticks()
    ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
    ax.set_xticks(viz.year.unique()) # https://stackoverflow.com/questions/21910986/why-set-xticks-doesnt-set-the-labels-of-ticks/21911813#21911813
```

```python
visualiseVotes(party='LIBERTARIAN')
```
