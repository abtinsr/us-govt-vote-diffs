```python
def votes_by_state_and_party(data, year):
    
    df = data[data['year']==year]
    
    df = (
        df.groupby(["state", "party"])
        .aggregate({"total_votes":"sum"})
        .sort_values(["state","total_votes"], ascending=False)
        .reset_index()
    )
    
    return(df)
```

```python
votes_by_state_and_party(data=senate, year=2018)
```

```python
def margins_by_state(data):
    
    df = votes_by_state_and_party(data)
    
    df = (
        df.pivot(index='state', 
                 columns='party', 
                 values='total_votes')
        .reset_index()
        .set_index('state')
    )
    
    df = df.div(df.sum(axis=1), axis=0) # https://stackoverflow.com/questions/50820659/compute-row-percentages-in-pandas-dataframe
    
    df = df[['REP', 'DEM']].reset_index()
    
    return(round(df,3))
```

```python
margins_by_state(pres).head(5)
```

```python
margins_by_state(senate).head(5)
```
